```python
import pandas as pd
```

```python
#Read in data from csv files
test_df = pd.read_csv(r"C:\Users\halcm\Desktop\ProjectData\tmdb-box-office-prediction\test.csv")

#Take a glimpse at the data
test_df.head()
```

```python
#Determine how many rows I'm working with for each column
test_df.count()
```

```python
#Drop unwanted columns from dataframe
test_df.drop(["belongs_to_collection",
          "homepage",
          "imdb_id",
          "original_language",
          "original_title",
          "overview",
          "poster_path",
          "spoken_languages",
          "status",
          "Keywords",
          "cast",
          "crew"
         ], axis = 1, inplace = True)
```

```python
test_df.count()
```

```python
#Remove missing values
test_df = test_df.dropna(how='any')

#See if removing the missing values evened out the data
test_df.count()
```

```python
#Rename columns to keep column headers less than 20 characters
test_df=test_df.rename(columns = {
    'id':'ID',
    'budget':'Budget',
    'genres':'Genres',
    'popularity':'Popularity',
    'production_companies':'Prod Companies',
    'production_countries':'Prod Countries',
    'release_date':'Release Date',
    'runtime':'Runtime',
    'tagline':'Tagline',
    'title':'Title',
    'revenue':'Revenue',
    })
test_df.head()
```

```python
#Reorder the columns in the dataframe
test_df = test_df[[
    'ID',
    'Title',
    'Release Date',
    'Budget',
    'Revenue',
    'Genres',
    'Popularity',
    'Runtime',
    'Tagline',
    'Prod Companies',
    'Prod Countries'
]]
test_df.head()
```
