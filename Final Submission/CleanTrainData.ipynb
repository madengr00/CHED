```python
# Check for nulls
train_df.loc[train_df["Release_Date"].isnull()]
# There is 1 movie w/o a release date. Looking it up in imdb by imdb_id = tt0210130, it was release in march 2000
train_df.loc[train_df["Release_Date"].isnull(), "Release_Date"] = "05/01/2000"
# Parse the string to a date
train_df["Release_Date"] = pd.to_datetime(train_df["Release_Date"])
# Create columns for each part of the date
train_df["Release_Date_weekday"] = train_df["Release_Date"].dt.weekday.astype(int)
train_df["Release_Date_month"] = train_df["Release_Date"].dt.month.astype(int)
train_df["Release_Date_year"] = train_df["Release_Date"].dt.year.astype(int)
# The year is formatted as yy as opposed to yyyy, and therefore the century is sometimes incorrect.
train_df["Release_Date_year"] = np.where(train_df["Release_Date_year"]>2019, train_df["Release_Date_year"]-100, train_df["Release_Date_year"])

# There are 4 movies from the 20th century that appear as 21st century due to the issue above. Fix them.
# train_df.loc[abs(train_df["Release_Date_year"]-train_df["imdb_year"])>2, "Release_Date_year"] = train_df["imdb_year"]
# drop_column("Release_Date")
```
